```python
# Library to generate random numbers
import random as rand

# Variable initialization
secret = rand.randint(1, 100)
count = 1

# While loop condition initialization
guess = int(input("Please enter your guess: "))

# Prompt until user finds the secret
while guess != secret:
  # Update the value of attempts
  count += 1
  # Check is guess is smaller or bigger than secret
  if guess < secret:
    print("Your guess is smaller than my secret number!")
  else:
    print("Your guess is greater than my secret number!")
  # While loop condition update
  guess = int(input("Please enter your guess: "))

# Display the result
print("You found it in", count, "attempt(s)!")
```
